## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Data Wrangling Code

# 1. Convert 'Date' columns to datetime format
# Convert 'Date' columns to datetime format with day first
features['Date'] = pd.to_datetime(features['Date'], dayfirst=True)
sales['Date'] = pd.to_datetime(sales['Date'], dayfirst=True)


# 2. Fill missing values in features dataset
# For MarkDown columns, fill missing values with 0 (assuming no markdowns applied)
markdown_cols = ['MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5']
features[markdown_cols] = features[markdown_cols].fillna(0)

# For CPI and Unemployment, fill missing values with forward fill method
features['CPI'] = features['CPI'].fillna(method='ffill')
features['Unemployment'] = features['Unemployment'].fillna(method='ffill')

# 3. Merge features and sales datasets on 'Store' and 'Date'
data_merged = pd.merge(sales, features, on=['Store','Date'], how='left')

# 4. Merge the stores dataset to include store type and size
data_merged = pd.merge(data_merged, stores, on='Store', how='left')

# 5. Reset index
data_merged.reset_index(drop=True, inplace=True)

# 6. Quick check of final dataset
print("Merged Dataset Shape:", data_merged.shape)
display(data_merged.head())
````

### What all manipulations have you done and insights you found?

**Data Manipulations:**
- Converted `Date` columns to datetime (`dayfirst=True`) for correct format.
- Filled missing values:
  - `MarkDown1–5` → 0 (no promotion applied)
  - `CPI` & `Unemployment` → forward fill
- Merged **sales + features + stores** datasets to create a unified dataset.
- Reset index for a clean structure.

**Insights:**
- Final dataset: 421,570 rows × 17 columns, combining sales, promotions, store info, and economic factors.
- Holiday weeks are clearly marked for analyzing impact on sales.
- Markdown columns are numeric and non-null, enabling promotion effect analysis.
- External factors (Temperature, Fuel_Price, CPI, Unemployment) aligned with sales dates.
- Store info (`Type`, `Size`) allows segmentation and performance comparison.
- Dataset is now ready for **EDA, visualization, and modeling**.

##### Why did you choose the specific statistical test?

- We are comparing two independent groups (holiday vs non-holiday weeks).

- The variable of interest (Weekly Sales) is continuous, making a t-test the right choice.

- Welch’s version of the t-test is more robust when variances differ.

### 1. Handling Missing Values

````python
# Handling Missing Values & Missing Value Imputation

# Check missing values
print("Missing values before imputation:\n")
print(data_merged.isnull().sum())

# 1. Fill markdown-related missing values with 0 (business logic: no markdown)
for col in ['MarkDown1','MarkDown2','MarkDown3','MarkDown4','MarkDown5']:
    data_merged[col] = data_merged[col].fillna(0)

# 2. Forward fill & backward fill for CPI and Unemployment (time-series continuity)
data_merged['CPI'] = data_merged['CPI'].fillna(method='ffill').fillna(method='bfill')
data_merged['Unemployment'] = data_merged['Unemployment'].fillna(method='ffill').fillna(method='bfill')

# 3. Double-check missing values after imputation
print("\nMissing values after imputation:\n")
print(data_merged.isnull().sum())
````

#### What all missing value imputation techniques have you used and why did you use those techniques?

MarkDown1–MarkDown5 → Missing values were replaced with 0.

Business logic: If no markdown/discount is recorded for that week, it means no promotion was applied, hence 0 is the most logical replacement.

CPI (Consumer Price Index) & Unemployment → Missing values were handled using forward fill (ffill) and backward fill (bfill).

Reason: These are time-series variables that usually change gradually over time. Carrying forward the most recent valid value (or using the next available value) ensures continuity and avoids introducing unrealistic jumps.

Other Columns → No missing values were found, so no imputation was required.

### 2. Handling Outliers

````python
# Handling Outliers & Outlier treatments

import numpy as np

# 1. Detect outliers using IQR method for Weekly_Sales
Q1 = data_merged['Weekly_Sales'].quantile(0.25)
Q3 = data_merged['Weekly_Sales'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Capping outliers instead of removing
data_merged['Weekly_Sales'] = np.where(
    data_merged['Weekly_Sales'] > upper_bound, upper_bound,
    np.where(data_merged['Weekly_Sales'] < lower_bound, lower_bound,
             data_merged['Weekly_Sales'])
)

# 2. Apply log transformation to skewed MarkDown columns
markdown_cols = ['MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4', 'MarkDown5']
for col in markdown_cols:
    # Add 1 to avoid log(0)
    data_merged[col] = np.log1p(data_merged[col])

# 3. Check remaining outliers in continuous variables (Temperature, Fuel_Price, CPI, Unemployment)
continuous_cols = ['Temperature', 'Fuel_Price', 'CPI', 'Unemployment']

# Boxplot check
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15,8))
for i, col in enumerate(continuous_cols, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(data=data_merged, y=col)
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

In our dataset, we detected and treated outliers using the following approaches:

IQR (Interquartile Range) Capping for Weekly_Sales:

Outliers in Weekly_Sales were detected using the IQR method (values beyond 1.5 × IQR from Q1 and Q3).

Instead of removing these records (which could cause data loss), we applied capping, i.e., replacing values beyond the upper bound with the upper threshold and values below the lower bound with the lower threshold.

This ensures that extreme fluctuations in sales are handled without dropping important observations.

Log Transformation for Skewed Variables (MarkDown1–MarkDown5):

These columns contained highly skewed distributions with large outliers.

To normalize their distribution and reduce the influence of extreme values, we applied log transformation (log1p).

This stabilizes variance and makes the data more suitable for modeling.

Visual Outlier Detection (Boxplots for Continuous Features):

We used boxplots to check for outliers in Temperature, Fuel_Price, CPI, and Unemployment.

These visualizations helped us confirm whether further treatment was necessary.

````python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Small set of alpha values to test quickly
alpha_values = [0.01, 0.1, 1, 10]

best_alpha = None
best_r2 = -np.inf
best_model = None

for alpha in alpha_values:
    ridge = Ridge(alpha=alpha, random_state=42)
    ridge.fit(X_train_fixed, y_train)
    y_pred = ridge.predict(X_test_fixed)
    r2 = r2_score(y_test, y_pred)

    if r2 > best_r2:
        best_r2 = r2
        best_alpha = alpha
        best_model = ridge

# Final predictions with best model
y_pred_best = best_model.predict(X_test_fixed)

# Evaluation
mae = mean_absolute_error(y_test, y_pred_best)
rmse = np.sqrt(mean_squared_error(y_test, y_pred_best))

print(f"Best Alpha: {best_alpha}")
print(f"Ridge Regression (Tuned) Performance:")
print(f"MAE: {mae:.2f}, RMSE: {rmse:.2f}, R²: {best_r2:.2f}")
````

### 1. Which Evaluation metrics did you consider for a positive business impact and why?

For this project, the chosen evaluation metrics are:

MAE (Mean Absolute Error):
Indicates the average magnitude of prediction errors. In business terms, it tells how much the model is deviating (on average) from actual weekly sales. Lower MAE means more reliable demand forecasting and helps in better inventory and staffing decisions.

RMSE (Root Mean Squared Error):
Penalizes larger errors more than MAE. In retail sales forecasting, avoiding big mistakes (e.g., underestimating holiday sales) is crucial. RMSE directly highlights whether the model avoids such critical mispredictions.

R² Score (Coefficient of Determination):
Measures how well the independent variables explain the variability in sales. Higher R² means better prediction reliability. For the business, this ensures that the chosen features (Markdowns, Fuel Price, CPI, Store Size, etc.) effectively capture sales behavior.
